```python
import numpy as np
import matplotlib.pyplot as plot
import os
import platform
```

Generate data from testing script.

```python
# Compile program with GCC
os.system('gcc -o test_pid pid.c pid.h test_pid.c')
# Run program
# Use \\ for backslash and add .exe for windows
# Use / and no extension for anything else
windows = platform.system() == 'Windows'
os.system(f'.{'\\' if windows else '/'}test_pid{'.exe' if windows else ''}')
```

Get data from CSV file.

```python
# Columns of data per simulation
c = 3
```

```python
data = np.genfromtxt('pid_data.csv', delimiter=',')
n = (np.shape(data)[1] - 1) // c
# Get rid of titles 
data = data[1:,:]
```

Get titles from first row of CSV file

```python
with open('pid_data.csv', 'r') as f:
    titles = f.readline().strip().split(',')
```

Plot data.

```python
for i in range(n):
    for j in range(c):
        index = 1 + i*c + j
        plot.plot(data[:,0], data[:,index], label=titles[index])

    plot.legend()
    plot.xlabel('Time')
    plot.ylabel('Value')
    plot.show()
```
